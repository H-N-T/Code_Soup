```python
import pandas as pd
import string

# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Get the filename of the uploaded file (assuming there's only one file)
input_csv = list(uploaded.keys())[0]

# Define a function to remove non-numeric characters (handles non-strings)
def remove_non_numeric(text):
  if pd.api.types.is_string_dtype(text):
    if len(text) >= 2 and text[-2].isalpha() and text[-1].isdigit():
      return text[:-1]  # Remove the last character
    else:
      return ''.join(c for c in text if c in string.digits)
  else:
    return ''.join(c for c in str(text) if c in string.digits)

# Import the string module
import string

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(input_csv)

# Apply the remove_non_numeric function to the 'Patent number' column (modify if different)
df['Patent number'] = df['Patent number'].apply(remove_non_numeric)

# Define output filename (replace 'output.csv' with your desired name)
output_csv = 'output.csv'

# Save the modified DataFrame to a new CSV file
df.to_csv(output_csv, index=False)

print(f"CSV data processed and saved to: {output_csv}")
```

```python
import pandas as pd
import csv
import string
# Upload the CSV file
from google.colab import files
uploaded = files.upload()

# Get the filename of the uploaded file (assuming there's only one file)
input_csv = list(uploaded.keys())[0]

# Define a function to remove trailing digits after letters
def remove_trailing_digits(text):
  if pd.api.types.is_string_dtype(text):
    for i in range(len(text) - 1, -1, -1):
      if text[i].isdigit() and text[i-1].isalpha():
        return text[:i]
    return text
  else:
    return ''.join(c for c in str(text) if c in string.digits)

# Import the string module
import string

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(input_csv)

# Apply the remove_trailing_digits function to the 'Patent number' column (modify if different)
df['Patent number'] = df['Patent number'].apply(remove_trailing_digits)

# Define output filename (replace 'output.csv' with your desired name)
output_csv = 'output.csv'

# Save the modified DataFrame to a new CSV file
df.to_csv(output_csv, index=False)

print(f"CSV data processed and saved to: {output_csv}")
```

```python
import glob
import os
import shutil
```

```python
#Generation of list containing .ab1 input sequence files
#replace LIB_NAME with your library name
LIB_NAME=glob.glob("/content/drive/MyDrive/Input_sequences/*.ab1")
```

```python
#Creation of new folder for the output sequences
os.makedirs("/content/drive/MyDrive/output_sequences",exist_ok=True)
```

```python
#Generation of new files excluding the Source Bioscience order number
#replace LIB_NAME with your library name from above
#replace [forward order number] with the forward order number
#replace [reverse order number] with the reverse order number
for file_name in LIB_NAME:
  destination="/content/drive/MyDrive/output_sequences/"+file_name.replace("[forward order number]","").replace("[reverse order number]","").split("/")[-1]
  shutil.copyfile(file_name,destination)
```

This notebook is designed to remove the order number from the forwards and reverse Source Bioscience sequence reads such that they can be more easily alligned on Benchling.



Move your .ab1 files to a folder on the gDrive, and add the path to the input_folder box, e.g.

Now add the Source Bioscience order numbers as they appears on your files including the following underscore, e.g.

606521901_

```python
library_name = "LIB064" # @param {type:"string"}
input_folder = "/content/drive/My Drive/fun_with_code/file_renamer/LIB064/" # @param {type:"string"}
output_folder = "/content/drive/My Drive/fun_with_code/file_renamer/LIB064_output/" # @param {type:"string"}
fwd_order_number = "606837101_" # @param {type:"string"}
rev_order_number = "606837102_" # @param {type:"string"}
```

```python
#Mount google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Import packages
import glob
import os
import shutil
```

```python
#Generation of list containing .ab1 input sequence files
library_name=glob.glob(input_folder + "/*.ab1")
```

```python
#Creation of new folder for the output sequences
os.makedirs(output_folder,exist_ok=True)
```

```python
#Copy files to output folder removing the order number
for file_name in library_name:
  destination=output_folder+file_name.replace(fwd_order_number,"").replace(rev_order_number,"").split("/")[-1]
  shutil.copyfile(file_name,destination)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = r'C:\Users\Henry\OneDrive - fabricnano.com\trip_data.csv'
data = pd.read_csv(file_path)

def plot_data(time_unit='Time (hours)', start_time=None, end_time=None):
    # Validate the time unit
    if time_unit not in data.columns:
        raise ValueError("Invalid time unit. Choose from 'Time (hours)', 'Time (minutes)', or 'Time (seconds)'.")

    # Filter the data based on the specified range
    time = data[time_unit]
    data_filtered = data

    if start_time is not None:
        data_filtered = data_filtered[time >= start_time]

    if end_time is not None:
        data_filtered = data_filtered[time <= end_time]

    # Plot the data
    plt.errorbar(data_filtered[time_unit], data_filtered['Mean'], yerr=data_filtered['SD'], fmt='o', ecolor='grey', elinewidth=1, capsize=3)
    plt.xlabel(f'Time ({time_unit})')
    plt.ylabel('Mean Value')
    plt.title('Mean Value over Time with Error Bars')
    plt.show()

# Example usage
plot_data(time_unit='Time (hours)', start_time=0, end_time=72)
```

```python
import cv2
import os

def split_video_to_frames(video_path, output_folder, skip_frames=0):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # Capture the video from the given path
    cap = cv2.VideoCapture(video_path)
    
    # Check if the video is opened successfully
    if not cap.isOpened():
        print("Error: Could not open video.")
        return
    
    frame_count = 0
    saved_frame_count = 0
    
    while True:
        # Read a frame from the video
        ret, frame = cap.read()
        
        # If the frame was read correctly, ret is True
        if not ret:
            break
        
        # Check if the current frame should be saved
        if frame_count % (skip_frames + 1) == 0:
            # Construct the filename for the frame
            frame_filename = os.path.join(output_folder, f'frame_{saved_frame_count:05d}.png')
            
            # Save the frame as an image file
            cv2.imwrite(frame_filename, frame)
            
            # Print the frame count to the console
            print(f'Saved {frame_filename}')
            
            # Increment the saved frame count
            saved_frame_count += 1
        
        # Increment the frame count
        frame_count += 1
    
    # Release the video capture object
    cap.release()
    print("Done!")

# Example usage
skip_frames = 0  # This will save every 3rd frame
video_path = r'C:\Users\Henry\Downloads\Clip (July 14 2024 at 156 PM).mp4'
output_folder = r'C:\Users\Henry\Downloads\package_swipe_1f'
split_video_to_frames(video_path, output_folder, skip_frames)


```

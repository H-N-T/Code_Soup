```python
from google.colab import files

# Upload the two fasta files
print("Upload the first fasta file:")
uploaded = files.upload()

# Assuming the first fasta file is named 'fasta1.fasta'
fasta1_filename = next(iter(uploaded))

print("Upload the second fasta file:")
uploaded = files.upload()

# Assuming the second fasta file is named 'fasta2.fasta'
fasta2_filename = next(iter(uploaded))

# Function to parse fasta files
def parse_fasta(filename):
    sequences = {}
    with open(filename, 'r') as file:
        sequence_id = None
        sequence = ''
        for line in file:
            if line.startswith('>'):
                if sequence_id:
                    sequences[sequence_id] = sequence
                sequence_id = line.strip()[1:]
                sequence = ''
            else:
                sequence += line.strip()
        if sequence_id:
            sequences[sequence_id] = sequence
    return sequences

# Parse fasta files
fasta1_sequences = parse_fasta(fasta1_filename)
fasta2_sequences = parse_fasta(fasta2_filename)

# Compare sequences
sequences_in_fasta1_only = {key: value for key, value in fasta1_sequences.items() if key not in fasta2_sequences}
sequences_in_fasta2_only = {key: value for key, value in fasta2_sequences.items() if key not in fasta1_sequences}

# Print sequences in one fasta file but not the other
print("Sequences in the first fasta file but not in the second:")
for sequence_id, sequence in sequences_in_fasta1_only.items():
    print(f">{sequence_id}")
    print(sequence)

print("\nSequences in the second fasta file but not in the first:")
for sequence_id, sequence in sequences_in_fasta2_only.items():
    print(f">{sequence_id}")
    print(sequence)
```

```python
# Function to save sequences as fasta file
def save_fasta(filename, sequences):
    with open(filename, 'w') as file:
        for sequence_id, sequence in sequences.items():
            file.write(f">{sequence_id}\n{sequence}\n")

# Save sequences in new fasta files
save_fasta('sequences_in_fasta1_only.fasta', sequences_in_fasta1_only)
save_fasta('sequences_in_fasta2_only.fasta', sequences_in_fasta2_only)
```

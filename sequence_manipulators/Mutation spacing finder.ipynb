```python
# Define a regular expression pattern to extract numbers
pattern = r'\d+'

# Define an empty list to store the extracted numbers
extracted_numbers = []
```

```python
#create a new dictionary to store results
result = []

# Iterate through each string in the list
for string in input:
    # Use regular expression to find all numbers in the string
    numbers = re.findall(pattern, string)
    # Convert the numbers from strings to integers and append to the list
    numbers = [int(num) for num in numbers]
    extracted_numbers.append(numbers)
    result.append(abs(numbers[0] - numbers[1]))
```

```python
# Convert each integer into a string
string_list = [str(item) for item in result]

# Convert the list into a format for Excel
result_out = '\n'.join(string_list)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Concentration/Intensity data
concentration_intensity_data = np.array([
    [100, 6076286],
    [50, 5003354.333],
    [25, 3987367.705],
    [12.5, 2900409.667],
    [6.25, 1749406.667],
    [3.125, 1139807.667],
    [1.5625, 637610],
    [0.78125, 351567.6667]
])

# Calibration data
calibrated_intensity = np.array([
    6460108, 6005427, 5562299, 5085100, 4606530, 4117386, 3655068, 3253062, 2855893, 2557846,
    2304315, 2087972, 1910872, 1751137, 1618512, 1504301, 390296, 383905, 375747, 366471, 359555,
    351881, 345048, 338864, 333260, 326220, 319525, 312209, 306536, 301359, 297136
])

# Extracting concentration and intensity from the calibration data
calibrated_concentration = []
calibrated_intensity_values = []
for intensity, data in zip(calibrated_intensity, concentration_intensity_data):
    calibrated_intensity_values.append(intensity)
    calibrated_concentration.append(data[0])

# Generate a calibration curve
calibration_curve = np.polyfit(calibrated_intensity_values, calibrated_concentration, 3)
calibration_function = np.poly1d(calibration_curve)

# Plotting the calibration curve
plt.figure(figsize=(10, 6))
plt.scatter(calibrated_intensity_values, calibrated_concentration, color='blue', label='Calibration Data')
plt.plot(calibrated_intensity_values, calibration_function(calibrated_intensity_values), color='red', label='Calibration Curve')
plt.xlabel('Intensity')
plt.ylabel('Concentration')
plt.title('Calibration Curve')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
